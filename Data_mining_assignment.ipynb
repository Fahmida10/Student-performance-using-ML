<tr>
<th style="text-align:justify;font-family: 'Times New Roman';"><font size="4">
    To analyse which student failed overall the 3 subject we created a status column 'status' </font></th>
<tr>

```python
# checking how many students have failed overall

df['status'] = df.apply(lambda x : 'Fail' if x['pass_math'] == 'Fail' or
                           x['pass_reading'] == 'Fail' or x['pass_writing'] == 'Fail'
                           else 'pass', axis = 1)

df['status'].value_counts(dropna = False).plot.pie(colors = ['grey', 'crimson'])
plt.title('overall results', fontweight = 30, fontsize = 20)
plt.xlabel('status')
plt.ylabel('count')
plt.show()
```

<tr>
<th style="text-align:justify;font-family: 'Times New Roman';"><font size="4">
    Next, to classify the grades we decided to create a column 'grade' to see which category did the student obtain by using their percentage <br> <br>
 <br>  0  - 40 marks : grade E <br>
 <br>41 - 60 marks : grade D <br>
 <br> 60 - 70 marks : grade C <br>
 <br> 70 - 80 marks : grade B <br>
 <br> 80 - 90 marks : grade A <br>
 <br>90 - 100 marks : grade A*</font></th>
<tr>

```python
def getgrade(percentage, status):
  if status == 'Fail':
    return 'E'
  if(percentage >= 90):
    return 'A*'
  if(percentage >= 80):
    return 'A'
  if(percentage >= 70):
    return 'B'
  if(percentage >= 60):
    return 'C'
  if(percentage >= 40):
    return 'D'
  else :
    return 'E'

df['grades'] = df.apply(lambda x: getgrade(x['percentage'], x['status']), axis = 1 )

df['grades'].value_counts()
```

```python
df['grades'].value_counts(dropna = False).plot.pie(colors = ['blue','red', 'green','yellow','orange','brown'], explode=[0.02,0.02,0.02,0.02,0.02,0.02], autopct='%1.1f%%', startangle=90)
plt.title('Grade Pie Chart', fontweight = 40, fontsize = 20)
plt.show()
```

The majority of students obtained a grade of B or C (51.2%), whereas A* and E grades are the lowest, which are 5.8% and 5.1% respectively.

```python
def train_and_evaluate_classifiers(X_train, y_train, X_test, y_test):
    classifiers = [
        ('Logistic Regression', LogisticRegression(max_iter=10000)),
        ('K-Nearest Neighbors', KNeighborsClassifier()),
        ('Decision Trees', DecisionTreeClassifier()),
        ('Random Forest', RandomForestClassifier()),
        ('Gradient Boosting', GradientBoostingClassifier()),
        ('Support Vector Machine', SVC(kernel='linear')),
        ('Naive Bayes', GaussianNB()),
        ('AdaBoost', AdaBoostClassifier()),
        ('Extra Trees Classifier', ExtraTreesClassifier()),
        ('XGBoost', xgb.XGBClassifier(use_label_encoder=False, eval_metric='mlogloss'))
    ]

    results = {}

    for name, classifier in classifiers:
        # Train the classifier
        classifier.fit(X_train, y_train)

        metrics_data = {}
        for dataset_name, X, y in [('Training', X_train, y_train), ('Testing', X_test, y_test)]:
            # Predict on the dataset
            y_pred = classifier.predict(X)

            # Get the confusion matrix
            cm = confusion_matrix(y, y_pred)

            # Get evaluation metrics
            accuracy = accuracy_score(y, y_pred)
            precision = precision_score(y, y_pred, average='weighted')
            recall = recall_score(y, y_pred, average='weighted')
            f1 = f1_score(y, y_pred, average='weighted')

            metrics_data[dataset_name] = {
                'Confusion Matrix': cm,
                'Accuracy': accuracy,
                'Precision': precision,
                'Recall': recall,
                'F1 Score': f1
            }

            print(f"\n{name} - {dataset_name}:")
            print("Confusion Matrix:\n", cm)
            print("Accuracy:", accuracy)
            print("Precision:", precision)
            print("Recall:", recall)
            print("F1 Score:", f1)

        results[name] = metrics_data

    return results

# Adjusting y labels
y_train_adjusted = y_train - 1
y_test_adjusted = y_test - 1

# Execute the function
evaluation_results = train_and_evaluate_classifiers(x_train, y_train_adjusted, x_test, y_test_adjusted)
```

**Hyper Parameter Tuning**

```python
# Reduced Hyperparameters grids
dt_params = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10, 50],
    'min_samples_split': [2, 10],
    'min_samples_leaf': [1, 4]
}

rf_params = {
    'n_estimators': [100, 200],
    'criterion': ['gini', 'entropy'],
    'max_features': ['auto'],
    'max_depth': [None, 20],
    'min_samples_split': [2, 10],
    'min_samples_leaf': [1, 4]
}

gb_params = {
    'n_estimators': [100, 200],
    'learning_rate': [0.01, 0.1],
    'max_depth': [3, 6],
    'subsample': [0.8, 1.0],
}

extra_trees_params = {
    'n_estimators': [100, 200],
    'criterion': ['gini', 'entropy'],
    'max_features': ['auto'],
    'max_depth': [None, 20],
    'min_samples_split': [2, 10],
    'min_samples_leaf': [1, 4]
}

# Tuning models (remains unchanged)
tuned_dt = tune_hyperparameters(DecisionTreeClassifier(), dt_params, x_train, y_train)
tuned_rf = tune_hyperparameters(RandomForestClassifier(), rf_params, x_train, y_train)
tuned_gb = tune_hyperparameters(GradientBoostingClassifier(), gb_params, x_train, y_train)
tuned_et = tune_hyperparameters(ExtraTreesClassifier(), extra_trees_params, x_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

def evaluate_model(model, X_train, y_train, X_test, y_test):
    # Predictions
    train_preds = model.predict(X_train)
    test_preds = model.predict(X_test)

    # Metrics
    metrics = {
        'Confusion Matrix (Train)': confusion_matrix(y_train, train_preds),
        'Accuracy (Train)': accuracy_score(y_train, train_preds),
        'Precision (Train)': precision_score(y_train, train_preds, average='weighted'),
        'Recall (Train)': recall_score(y_train, train_preds, average='weighted'),
        'F1 Score (Train)': f1_score(y_train, train_preds, average='weighted'),

        'Confusion Matrix (Test)': confusion_matrix(y_test, test_preds),
        'Accuracy (Test)': accuracy_score(y_test, test_preds),
        'Precision (Test)': precision_score(y_test, test_preds, average='weighted'),
        'Recall (Test)': recall_score(y_test, test_preds, average='weighted'),
        'F1 Score (Test)': f1_score(y_test, test_preds, average='weighted')
    }

    return metrics

# Assuming that you have your test dataset split as x_test and y_test.
models = {'DecisionTree': tuned_dt, 'RandomForest': tuned_rf, 'GradientBoosting': tuned_gb, 'ExtraTrees': tuned_et}

for name, model in models.items():
    print(f"Evaluating {name}:")
    results = evaluate_model(model, x_train, y_train, x_test, y_test)
    for metric_name, metric_value in results.items():
        if "Confusion" in metric_name:
            print(f"{metric_name}:\n{metric_value}")
        else:
            print(f"{metric_name}: {metric_value:.4f}")
    print("------------------------------------------------------")
```

**Hybrid Model**

```python
from sklearn.ensemble import VotingClassifier

# Create a hybrid model using the tuned models
hybrid_model = VotingClassifier(estimators=[
    ('dt', tuned_dt),
    ('rf', tuned_rf),
    ('gb', tuned_gb),
    ('et', tuned_et)
], voting='hard')

# Fit the hybrid model on the training data
hybrid_model.fit(x_train, y_train)

# Evaluate the hybrid model
print("Evaluating Hybrid Model:")
results = evaluate_model(hybrid_model, x_train, y_train, x_test, y_test)
for metric_name, metric_value in results.items():
    if "Confusion" in metric_name:
        print(f"{metric_name}:\n{metric_value}")
    else:
        print(f"{metric_name}: {metric_value:.4f}")
print("------------------------------------------------------")
```

<td style="text-align:justify;font-family: 'Times New Roman';"><font size="4">
The second limitation that occurred in our assignment is that our target variable is not the best dependent variable to measure student performance. This is because we only took math score, reading score, and also writing score as our y variable. To measure the performance of a student, the test scores are far from enough. Not all students can perform well in the study, however, some perform better in physical activities like sports, arts, music, etcetera. Therefore, the final analysis may not be that accurate.
<tr>
<td style="text-align:justify;font-family: 'Times New Roman';"><font size="4">
In the future, our team would like to investigate and analyze the dataset's characteristics in greater depth in order to extract the best features and improve accuracy. Furthermore, we would like to experiment with different algorithms as well as design and come up with our own algorithm that can reliably forecast for this dataset.


# Conclusion

<tr>
<td style="text-align:justify;font-family: 'Times New Roman';"><font size="4">The goal of this task was to determine if students’ performance in exams could be predicted based on the attributes and data provided by the Kaggle “Student Performance in Exams” dataset. Using the score method for validation, the accuracy of the model obtained is 82.8% for logistic regression, 87.6% for post-tuned logistic regression, and 100% for random forest. Therefore, it can be concluded that random forest is the best model for predicting student performance.
<tr>
<td style="text-align:justify;font-family: 'Times New Roman';"><font size="4">
From our work, there are limitations that can be seen. For example, we have only used two models to examine the attributes of the data. In the future, other models such as artificial neural network could be used to study the attributes in depth. Moreover, the attributes in the dataset we have chosen are limited, when in reality there are many more factors which may contribute to a student’s performance in exams such as learning environment, interactivity during classes, and individual personality.
<tr>
<td style="text-align:justify;font-family: 'Times New Roman';"><font size="4">
Our model can be said to have achieved its goal to a certain extent. The data mining techniques that we have used have allowed us to predict students’ performance on tests. However, further improvements can be made in order to build a more complete model which includes many more attributes.
